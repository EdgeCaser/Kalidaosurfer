```python
crunch_data = db.query("select "
                           "proposals_id "
                       "    ,proposals_dao_token_name "
                           ",proposals_votes_voter "
                           ",proposals_votes_vote"
                           ",proposals_votes_weight "

                           ",sum(proposals_votes_weight::float) over (Partition by proposals_id  order by proposals_votes_weight::float desc, proposals_votes_voter  asc) as cumulative_vp "

                       "    ,sum(proposals_votes_weight::float) over (Partition by proposals_id) as total_vp "
                           ",(proposals_votes_weight::float/sum(proposals_votes_weight::float) over (Partition by proposals_id)) as percentange_of_total_vp "

                       "    ,((sum(proposals_votes_weight::float) over (Partition by proposals_id  order by proposals_votes_weight::float desc, proposals_votes_voter asc))::float/sum(proposals_votes_weight::float) over (Partition by proposals_id)) as cum_percentage_of_total_vp "

                           ",round((sum(proposals_votes_weight::float) over (Partition by proposals_id  order by proposals_votes_weight::float desc, proposals_votes_voter asc))::float/sum(proposals_votes_weight::float) over (Partition by proposals_id)) as cum_percentage_of_total_vp_stepped "

                       "    ,row_number() over (Partition by proposals_id order by proposals_votes_weight::float desc, proposals_votes_voter asc) as proposal_voter_rank "

                           ",count(proposals_votes_voter) over (Partition by proposals_id  order by proposals_votes_weight::float desc, proposals_votes_voter asc) total_voters "

                           ",(count(*) over (Partition by proposals_id  order by proposals_votes_weight::float desc, proposals_votes_voter  asc))::float/(count(*) over (Partition by proposals_id))::float percentage_voters_counted "

                       ",round(100*(count(*) over (Partition by proposals_id  order by proposals_votes_weight::float desc, proposals_votes_voter asc))::float/(count(*) over (Partition by proposals_id)))::float percentage_voters_counted_stepped "
                       "from "
                       "    governance_data  "
                           ""
                       "Group by "
                           "proposals_id "
                       "    ,proposals_dao_token_name "
                           ",proposals_votes_voter "
                           ",proposals_votes_vote"
                           ",proposals_votes_weight "
                           ""
                       "Order by "
                       "    proposals_id, "
                       "    proposals_votes_weight::float desc").df()
crunch_data.drop_duplicates()
crunch_data.head(10)
```

```python
leader_ranks = db.query("with leader_ranks as "
                            "(Select distinct "
                            "   B.proposals_id"
                            "   ,B.proposals_votes_voter "
                            "   ,B.proposal_voter_rank "
                            "   ,(B.proposal_voter_rank +1) as leader_rank "
                            "From "
                            "   (select "
                                     "proposals_id "
                               "    ,proposals_dao_token_name "
                                   ",proposals_votes_voter "
                                   ",proposals_votes_vote"
                                   ",proposals_votes_weight "
                                   ",sum(proposals_votes_weight::float) over (Partition by proposals_id  order by proposals_votes_weight::float desc, proposals_votes_voter  asc) as cumulative_vp "
                               "    ,sum(proposals_votes_weight::float) over (Partition by proposals_id) as total_vp "
                                   ",(proposals_votes_weight::float/sum(proposals_votes_weight::float) over (Partition by proposals_id)) as percentange_of_total_vp "
                               "    ,((sum(proposals_votes_weight::float) over (Partition by proposals_id  order by proposals_votes_weight::float desc, proposals_votes_voter asc))::float/sum(proposals_votes_weight::float) over (Partition by proposals_id)) as cum_percentage_of_total_vp "
                                   ",round((sum(proposals_votes_weight::float) over (Partition by proposals_id  order by proposals_votes_weight::float desc, proposals_votes_voter asc))::float/sum(proposals_votes_weight::float) over (Partition by proposals_id)) as cum_percentage_of_total_vp_stepped "
                               "    ,row_number() over (Partition by proposals_id order by proposals_votes_weight::float desc, proposals_votes_voter asc) as proposal_voter_rank "
                                   ",count(proposals_votes_voter) over (Partition by proposals_id  order by proposals_votes_weight::float desc, proposals_votes_voter asc) total_voters "
                                   ",(count(*) over (Partition by proposals_id  order by proposals_votes_weight::float desc, proposals_votes_voter  asc))::float/(count(*) over (Partition by proposals_id))::float percentage_voters_counted "
                               ",round(100*(count(*) over (Partition by proposals_id  order by proposals_votes_weight::float desc, proposals_votes_voter asc))::float/(count(*) over (Partition by proposals_id)))::float percentage_voters_counted_stepped "
                                "from "
                                "    governance_data  "
                                ""
                                "Group by "
                                   "proposals_id "
                               "    ,proposals_dao_token_name "
                                   ",proposals_votes_voter "
                                   ",proposals_votes_vote"
                                   ",proposals_votes_weight "
                                ""
                                "Order by "
                                "    proposals_id, "
                                "    proposals_votes_weight desc) B "
                                "where "
                                "   B.cum_percentage_of_total_vp<=0.5) "
                            ""
                            "Select "
                            "   A.* "
                        "       ,B.leader_rank "
                            "From crunch_data A"
                            "   Join leader_ranks B on A.proposal_voter_rank = B.leader_rank and A.proposals_id = B.proposals_id"
                            ""
                            ).df()
leader_ranks.drop_duplicates()


leader_ranks.head(100)
```

The chart above shows what % of all possible votes has been cast (Y axis) as each incremental percent of the voting population casts their vote (X axis). Each line is a Proposal and has a unique color, so that a dot on each percent point represents what % of total voting power was accumulated by that group. The color represents which vote was cast.
The Orange X shows the average % of power accumulated across all elections.
